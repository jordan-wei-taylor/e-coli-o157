```python
from   matplotlib import pyplot as plt
from   genolearn.metrics import Metrics
from   genolearn.dataloader import DataLoader
import numpy as np
import os

path   = 'O157-region-subset'
models = 'lr', 'rf'
years  = range(2014, 2019)

color  = plt.rcParams['axes.prop_cycle'].by_key()['color']

M      = [100, 1000, 10000, 100000, 1000000]

dataloader = DataLoader('data', 'raw-data/meta-data.csv', 'Accession', 'Region', 'Year')
```

```python
info = {}
for model in models:

    info[model]   = {}
    recall        = info[model]['recall'] = {}
    times         = info[model]['time']   = {}
    for year in years:

        npz                    = np.load(os.path.join(path, str(year), model, 'results.npz'), allow_pickle = True)
        training_targets       = dataloader.load_Y(*npz['identifiers'])

        recall[year]           = Metrics(npz['target'], npz['predict'], 'recall')()

        times[year]            = npz['time']

        info[year]             = {}
        info[year]['counts']   = dict(zip(*np.unique(training_targets, return_counts = True)))
        info[year]['prop']     = {region : info[year]['counts'][region] / sum(info[year]['counts'].values()) for region in info[year]['counts']}

    info[model]['params'] = {key : npz[key] for key in list(npz)[-2:]}
    info[model]['x'] = npz[list(npz)[-2]]
    
info['counts'] = dict(zip(*np.unique(npz['target'], return_counts = True)))
info['prop']   = {region : info['counts'][region] / sum(info['counts'].values()) for region in info['counts']}
```

```python
kwargs = dict(va = 'center', xycoords = 'axes fraction')

for model in models:
    fig, ax = plt.subplots(8, 5, figsize = (20, 20), sharex = True, sharey = True)

    for j, year in enumerate(reversed(years)):
        for i, region in enumerate(info[model]['recall'][year]):
            x = info[model]['x']
            for k, result in enumerate(info[model]['recall'][year][region]):
                ax[i,j].plot(range(len(x)), result, c = color[k], alpha = 0.1)
                ax[i,j].plot(range(len(x)), result.mean(axis = -1), c = color[k])

            ax[i,j].set_xticks(range(0, len(x), 2))
            ax[i,j].set_xticklabels(x[::2])
            ax[i,j].set_ylim((-0.05, 1.05))
            ax[i,j].grid(ls = (0, (5, 5)))
            ax[i,j].annotate(f'{info[year]["counts"][region]:,d}', (0.06, 0.9), ha = 'left', **kwargs)
            ax[i,j].annotate(f'({info[year]["prop"][region] * 100:.1f})', (0.94, 0.9), ha = 'right', **kwargs)
            if j == 0:
                ax[i,j].set_ylabel(f'{region} : {info["counts"][region]:,d} ({info["prop"][region] * 100:.1f})')
        ax[0,j].set_title(f'{year}-2018')

    plt.show()
    
```
